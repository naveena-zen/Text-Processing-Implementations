```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

doc1 = "every challenge in life makes us stronger"
doc2 = "life challenges help people become stronger"

vec = CountVectorizer().fit_transform([doc1, doc2])
similarity = cosine_similarity(vec)[0][1]

print("Cosine Similarity:", similarity)
```

```python
doc1 = set("every challenge in life makes us stronger".split())
doc2 = set("life challenges help people become stronger".split())

similarity = len(doc1 & doc2) / len(doc1 | doc2)

print("Jaccard Similarity:", similarity)
```
